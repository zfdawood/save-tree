```python
# create dataset from the files in the data folder
dataset = None
for f in os.listdir("./data"):
  dataset = pd.concat([dataset, pd.read_csv(f"./data/{f}")])
dataset
```

```python
print(dataset.shape)
```

```python
# boxplot of the times (in milliseconds) comparing using the interaction technique
dataset.boxplot(column=["time"], by="tool")
```

```python
#get "speed" - size of characters typed (3 times the original text for three texts) divided by total time taken
# roughly normalize to a  words per minute value to better compare across using the tool and not using the tool
dataset["speed"] = 3/5 * dataset["original"].str.len() / (dataset["time"] / 60000)

# get means of the numerical columns and compare them by whether the tool was used or not for that trial
dataset.groupby("tool").mean(numeric_only=True)
```

```python
# get medians too
dataset.groupby("tool").median(numeric_only=True)
```

```python
# boxplot of errors
dataset.boxplot(column=["errors"], by="tool")
```
